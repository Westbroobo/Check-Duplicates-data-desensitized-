```python
import pandas as pd
```

```python
df['Description'] = df['Description'].str.split('\n')
df = df.explode('Description')
df = df.drop_duplicates(keep='first')
df.reset_index(inplace=True,drop=True)
# df_gb = df.groupby(['Part Number'],as_index=False).agg({"Engine":";".join})
```

```python
df
```

```python
df['SKU'] = df['SKU'].astype(str)
```

```python
df['Description'] = df['Description'].apply(lambda x: x.strip())
```

```python
df_gb = df.groupby('Description',as_index=False,sort=False).agg({'SKU':','.join})
```

```python
df_gb['goods_num'] = df_gb['SKU'].apply(lambda x: len(set(x.split(','))))
```

```python
df_gb
```

```python
df_gb['list_vehicle'] = df_gb['Description'].apply(lambda x: x.split(' '))
df_gb
```

```python
df_gb['year'] = df_gb['list_vehicle'].apply(lambda x: x[-1] if len(x)>2 else '')
df_gb['make'] = df_gb['list_vehicle'].apply(lambda x: ' '.join(x[0:2]) if x[0]=='Land' or x[0]=='Alfa' else x[0])
df_gb['make'] = df_gb['make'].apply(lambda x: x.strip())
df_gb['model'] = df_gb['list_vehicle'].apply(lambda x: (' '.join(x[2:-1]) if x[0]=='Land' or x[0]=='Alfa' else ' '.join(x[1:-1])) if len(x)>2 else x[-1])
df_gb['model'] = df_gb['model'].apply(lambda x: x.strip())
df_gb
```

```python
def create_year_range(year_str):
    if not year_str:
        return list()
    try:
        parts = year_str.split('-')
        start = int(parts[0])
        end = int(parts[1]) if len(parts) > 1 else start
        return list(range(start, end + 1))
    except (ValueError, IndexError):
        return []  
```

```python
df_gb['list_year'] = [create_year_range(year) for year in df_gb['year']]
df_gb
```
